```python
import numpy as np
```

```python
with open("/home/curt/Downloads/input15.txt", newline="\n") as f:
    raw_data = f.read().splitlines()

risk = np.array([[int(c) for c in l] for l in raw_data])
```

```python
lowest = risk.diagonal().sum() + risk.diagonal(1).sum() - risk[0,0]

def get_possible_next(x,y,path):
    possible, m = [], risk.shape[0]-1
    if x<m and ((np:=(x+1,y)) not in path): possible.append(np)
    if y<m and ((np:=(x,y+1)) not in path): possible.append(np)
    #if x>0 and ((np:=(x-1,y)) not in path): possible.append(np)
    #if y>0 and ((np:=(x,y-1)) not in path): possible.append(np)
    return possible


def next_step(cur_pos, path, cur_total):
    global lowest
    for next_pos in get_possible_next(*cur_pos, path):
        if (next_total:=(cur_total + risk[next_pos])) < lowest:
            if list(next_pos)==[c-1 for c in risk.shape]: 
                lowest=next_total
                print(lowest)
            else: next_step(next_pos, path+[next_pos], next_total)
        

print(lowest)
next_step((0,0), [], 0)
print(lowest)
```
